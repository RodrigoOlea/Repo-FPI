## 1. Operadores booleanos: respuestas verdaderas o falsas

Un booleano es un tipo de dato que solo puede tener uno de dos valores: `True` (verdadero) o `False` (falso). Podemos pensarlo como la respuesta a una pregunta que solo admite «sí» o «no».

Python usa los siguientes operadores para hacer comparaciones:

- `>`: mayor que
- `>=`: mayor o igual que
- `<`: menor que
- `<=`: menor o igual que
- `==`: igual a
- `!=`: distinto de

Cada comparación produce `True` o `False` como respuesta. Más adelante usaremos estas respuestas para decidir **qué instrucciones ejecutar** con `if` y **cuándo seguir repitiendo instrucciones** con `while`.

```python
# Objetivo: comprobar que toda comparación responde con True o False.
print("¿2.0 es mayor que 5?", 2.0 > 5)
print("¿8 es mayor o igual que 3.7?", 8 >= 3.7)
print("¿1.5 es menor que 1.4?", 1.5 < 1.4)

# Conclusión: aunque comparamos números, cada resultado es un booleano.
```

**Error común:**

- Confundir `=` con `==`. Un solo signo igual le **asigna** un valor a una variable, por ejemplo `edad = 20`. Dos signos iguales **comparan** valores, por ejemplo `edad == 20`.

## 2. Operadores lógicos: **combinar condiciones**

Los operadores lógicos permiten unir o modificar condiciones:

- **`and` (y):** el resultado es `True` solo cuando ambas condiciones son verdaderas.
- **`or` (o):** el resultado es `True` cuando al menos una condición es verdadera.
- **`not` (no):** invierte una condición; `True` pasa a ser `False` y viceversa.

```python
# Objetivo: observar cómo and, or y not combinan o modifican condiciones.

# AND: para conducir deben cumplirse los dos requisitos: ser mayor de edad y tener licencia.
edad = 21
tiene_licencia = True
puede_conducir = edad >= 18 and tiene_licencia
print("¿Puede conducir?", puede_conducir)
# Conclusión: el resultado es True porque es mayor de edad y tiene licencia.

# OR: basta con pertenecer a uno de los dos grupos.
es_docente = False
es_estudiante = True
puede_entrar = es_docente or es_estudiante
print("¿Puede entrar a la universidad?", puede_entrar)
# Conclusión: el resultado es True porque al menos una condición es verdadera.

# NOT: negamos el estado de la puerta.
puerta = False # La puerta esta cerrada
puerta = not puerta # Negamos el booleano -> Abrimos la puerta
print("¿La puerta está abierta?", puerta) 
# Conclusión: not False produce True.
```

## 3. Sentencias condicionales: **tomar decisiones**

Una sentencia condicional permite que un programa elija qué hacer. Python evalúa una condición que produce `True` o `False` y, a partir de ese resultado, decide qué bloque de instrucciones ejecutar.

La sangría —los espacios al comienzo de una línea— indica qué instrucciones pertenecen a cada bloque.

### 3.1 `if`: ejecutar algo solo si se cumple una condición

`if` significa «si». El bloque bajo el `if` se ejecuta **únicamente cuando su condición es verdadera**. Si es falsa, Python salta ese bloque y continúa con el resto del programa.

```python
'''
Estructura general:

if condicion:
    instrucciones que se ejecutan si la condición es True
'''


# Objetivo: mostrar un mensaje solamente cuando el número sea par.
numero = 5

if numero % 2 == 0:  # Un número es par si el resto que queda al dividirlo por 2 es 0.
    print("El número es par.")

# Conclusión: como numero vale 5, la condición es False y el bloque no se ejecuta.
```

### 3.2 `if-else`: elegir entre dos caminos

`if-else` cubre dos posibilidades. Si la condición es verdadera, se ejecuta el bloque de `if`. Si es falsa, se ejecuta el bloque de `else`. Por lo tanto, siempre se elige uno de los dos caminos.

```python
'''
Estructura general:

if condicion:
    instrucciones si la condición es True
else:
    instrucciones si la condición es False
'''

# Objetivo: clasificar un número en una de dos categorías.
numero = 7

if numero >= 0:
    print("El número es positivo o cero.")
else:
    print("El número es negativo.")

# Conclusión: siempre se imprime un solo mensaje, porque solo hay dos caminos.
```

### 3.3 `if-elif-else`: elegir entre tres o más caminos

`elif` es una forma corta de escribir *else if*, que podemos leer como «si no, revisa si...». No inicia una decisión independiente: pertenece a la misma cadena que el `if` anterior.

Python revisa una cadena `if-elif-else` de arriba hacia abajo:

1. Primero revisa la condición del `if`.
2. Si es verdadera, ejecuta ese bloque y omite todos los `elif` y el `else` restantes.
3. Si es falsa, revisa el primer `elif`. El proceso continúa hasta encontrar una condición verdadera.
4. Si ninguna condición es verdadera, ejecuta el `else`.

Por eso, en una cadena `if-elif-else` **se ejecuta como máximo un bloque**. Esta estructura sirve para casos que queremos tratar como alternativas: si ocurre uno, los demás ya no deben ocurrir.

Esto es diferente de escribir varios `if` seguidos. Cada `if` abre una decisión nueva e independiente, por lo que Python revisa todas sus condiciones. Si dos o tres son verdaderas, se pueden ejecutar dos o tres bloques. Conviene usar varios `if` cuando las situaciones pueden ocurrir al mismo tiempo.

Finalmente, un `if` puede aparecer solo: no estamos obligados a indicar qué hacer cuando su condición es falsa. En cambio, un `else` nunca puede aparecer solo, porque significa «en el caso contrario» y necesita estar asociado a un `if`. Del mismo modo, un `elif` siempre necesita un `if` anterior.

```python
'''
Estructura general de una cadena:

if primera_condicion:
    instrucciones del primer caso
elif segunda_condicion:
    instrucciones del segundo caso
else:
    instrucciones si ninguna condición anterior se cumple
'''

# Objetivo: comparar una cadena if-elif con varios if independientes.
puntaje = 95

# Esta cadena elige una sola categoría.
if puntaje >= 90:
    print("Categoría: desempeño destacado")
elif puntaje >= 60:
    print("Categoría: desempeño aprobado")
else:
    print("Categoría: desempeño no aprobado")

# Estos if son independientes y revisan todas las afirmaciones.
if puntaje >= 60:
    print("El puntaje cumple el requisito de aprobación.")
if puntaje >= 90:
    print("El puntaje cumple el requisito de distinción.")
if puntaje == 100:
    print("El puntaje es el máximo posible.")

'''
Conclusión: if-elif imprime una sola categoría. Los if independientes pueden
imprimir varios mensajes porque cada condición se revisa por separado.
'''
```

**Errores comunes:**

- Usar una sangría distinta dentro del mismo bloque. Lo más habitual es usar cuatro espacios.
- Olvidar los dos puntos (`:`) al final de una línea con `if`, `elif` o `else`.
- Ordenar mal condiciones que se superponen. En una cadena `if-elif`, conviene revisar primero el caso más específico, porque Python omite los casos restantes después de encontrar el primero verdadero.

## 4. Bucle `while`: repetir mientras se cumpla una condición

`while` revisa una condición y repite un bloque de código **mientras esa condición sea verdadera**. Antes de cada nueva repetición, vuelve a revisar la condición.

Esta forma de **repetir** es especialmente útil cuando no sabemos de antemano cuántas repeticiones harán falta. Por ejemplo, podemos duplicar una cantidad hasta que alcance una meta, sin calcular previamente cuántas duplicaciones necesitaremos.

```python
'''
Estructura general:

while condicion:
    instrucciones que se repiten mientras la condición sea True
'''

# Objetivo: duplicar una cantidad hasta alcanzar una meta.
cantidad = 2
meta = 100
duplicaciones = 0

while cantidad < meta:
    cantidad = cantidad * 2
    duplicaciones = duplicaciones + 1

print("Cantidad final:", cantidad)
print("Número de duplicaciones:", duplicaciones)
# Conclusión: while se detiene cuando cantidad deja de ser menor que meta.
```

**Errores comunes:**

- Crear una condición que nunca llegue a ser falsa. En ese caso, el programa no deja de repetir el bloque: esto se llama **bucle infinito**.
- Olvidar actualizar dentro del bucle el dato que participa en la condición. Si ese dato no cambia, la condición podría mantenerse verdadera para siempre.

## 5. Bucle `for`: recorrer elementos uno por uno

`for` toma, en orden, cada elemento de una **colección**, como un texto, una lista o una secuencia creada con `range()`. En cada repetición, la variable ubicada a la izquierda de `in` guarda el elemento actual de la coleccion. Esta variable es nuestro **iterador**.

Usamos `for` cuando queremos realizar una acción una vez por cada elemento.

```python
'''
Estructura general:

for elemento in coleccion:
    instrucciones que se ejecutan una vez por cada elemento
'''

# Objetivo: observar que un texto puede recorrerse letra por letra.
texto = "Python"

for caracter in texto:
    print(caracter)

# Conclusión: el bloque se ejecuta seis veces, una por cada carácter de "Python".
```

```python
# Objetivo: observar que for también puede recorrer una lista.
frutas = ["manzana", "pera", "naranja"]

for fruta in frutas:
    print("Fruta actual:", fruta)

# Conclusión: fruta toma un valor distinto de la lista en cada repetición.
```

## 6. `range()`: crear secuencias de números para un `for`

`range()` crea una secuencia de números enteros que `for` puede recorrer. Se puede usar de tres formas:

- `range(fin)` comienza en `0` y se detiene antes de `fin`.
- `range(inicio, fin)` comienza en `inicio` y se detiene antes de `fin`.
- `range(inicio, fin, paso)` avanza según el `paso`. El paso puede ser positivo o negativo, pero no puede ser `0`.

El valor de `fin` nunca se incluye. Por ejemplo, `range(1, 4)` produce `1`, `2` y `3`.

```python
# Objetivo: comparar las distintas formas de construir una secuencia con range().

print("range(5): comienza en 0")
for numero in range(5):
    print(numero)  # Imprime 0, 1, 2, 3 y 4; nunca imprime 5.

print("range(3, 7): comienza en 3")
for numero in range(3, 7):
    print(numero)  # Imprime 3, 4, 5 y 6; nunca imprime 7.

print("range(0, 10, 2): avanza de dos en dos")
for numero in range(0, 10, 2):
    print(numero)  # Imprime los números pares desde 0 hasta 8.

print("range(5, 0, -1): retrocede de uno en uno")
for numero in range(5, 0, -1):
    print(numero)  # Imprime 5, 4, 3, 2 y 1; nunca imprime 0.

# Conclusión: inicio se incluye, fin no se incluye y paso controla el avance.
```

**Error común:**

- Esperar que `range()` incluya el valor final. Este descuido suele producir una repetición menos o una más de la necesaria. Por ejemplo, para recorrer del `1` al `10`, debemos escribir `range(1, 11)`.

## 7. ¿Cuándo usar `while` y cuándo usar `for`?

Los dos bucles repiten instrucciones, pero se adaptan a situaciones distintas:

- Usa **`for`** cuando puedas identificar los elementos que recorrerás o sepas cuántas repeticiones necesitas. Por ejemplo: revisar cada nombre de una lista o repetir algo diez veces.
- Usa **`while`** cuando la repetición dependa de una condición y no sepas cuántos intentos serán necesarios. Por ejemplo: seguir preguntando hasta recibir una respuesta válida.

Una pregunta práctica ayuda a elegir: «¿Puedo enumerar las repeticiones antes de comenzar?». Si la respuesta es sí, normalmente conviene `for`. Si la respuesta es no, normalmente conviene `while`.

```python
# Objetivo: resolver la misma tarea con for y while para comparar ambos bucles.
n = 6

# Con for recorremos directamente los números desde 1 hasta n.
suma_for = 0
for numero in range(1, n+1):  # Usamos n + 1 porque range no incluye el final.
    suma_for = suma_for + numero

# Con while debemos crear y actualizar manualmente el número actual.
suma_while = 0
numero = 1
while numero <= n:
    suma_while = suma_while + numero
    numero = numero + 1

print("Suma con for:  ", suma_for)
print("Suma con while:", suma_while)

'''
Conclusión: ambos entregan el mismo resultado, pero for expresa mejor esta tarea
porque conocemos de antemano todos los números que queremos recorrer.
'''
```
