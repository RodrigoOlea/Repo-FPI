## 1. Funciones propias y su sintaxis

Una **función propia** es un conjunto de instrucciones al que le damos un nombre. Después de definirla, podemos ejecutar esas instrucciones todas las veces que necesitemos llamando a la función por su nombre.

Las funciones sirven para:

- Evitar repetir las mismas instrucciones.
- Dividir un problema grande en tareas pequeñas.
- Dar un nombre claro a cada tarea del programa.
- Usar una misma solución con distintos datos.

La **sintaxis** es la forma en que debemos escribir el código para que Python lo entienda. Para definir una función usamos la palabra `def`. Para llamarla, escribimos su nombre seguido de paréntesis.

```python
'''
Estructura general:

def nombre_funcion(parametros):
    instrucciones de la función
'''

# Objetivo: definir una función sencilla y llamarla más de una vez.
def mostrar_bienvenida():
    print("¡Bienvenida al curso de programación!")

mostrar_bienvenida()
mostrar_bienvenida()

# Conclusión: definimos las instrucciones una vez, pero podemos ejecutarlas
# tantas veces como llamemos a la función.
```

### 1.1 Partes de una función

En `def mostrar_bienvenida():` podemos reconocer cuatro partes:

1. `def` indica que comenzaremos a definir una función.
2. `mostrar_bienvenida` es el nombre elegido para la función.
3. Los paréntesis guardarán los parámetros que la función necesite. En este primer ejemplo están vacíos.
4. Los dos puntos (`:`) abren el cuerpo de la función. Sus instrucciones llevan cuatro espacios de sangría.

Escribir solamente `mostrar_bienvenida` se refiere a la función, pero no la ejecuta. Los paréntesis de `mostrar_bienvenida()` son los que realizan la llamada.

## 2. Parámetros y argumentos

Los **parámetros** son nombres que aparecen en la definición y representan los datos que la función necesitará. Funcionan como espacios que completaremos al llamar a la función.

Los **argumentos** son los valores concretos que entregamos durante cada llamada.

Por ejemplo, en `def mostrar_area(base, altura):`, `base` y `altura` son parámetros. En `mostrar_area(5, 3)`, `5` y `3` son argumentos.

```python
# Objetivo: reutilizar una función entregándole argumentos diferentes.
def mostrar_area(base, altura):
    area = base * altura
    print("Área del rectángulo:", area)

mostrar_area(5, 3)
mostrar_area(8, 2)

# Conclusión: la función realiza el mismo procedimiento en ambas llamadas,
# pero sus parámetros reciben valores distintos.
```

**Errores comunes:**

- Entregar menos o más argumentos de los que la función necesita.
- Olvidar la sangría del cuerpo de la función.
- Olvidar los dos puntos (`:`) al final de la línea con `def`.
- Usar nombres poco claros. `calcular_area` comunica mejor su propósito que un nombre como `hacer_cosa`.

## 3. Momento de definición y momento de ejecución

Definir y ejecutar una función son momentos distintos:

- **Momento de definición:** Python encuentra `def`, guarda el nombre y las instrucciones de la función, pero no ejecuta su cuerpo.
- **Momento de ejecución:** Python encuentra una llamada, asigna los argumentos a los parámetros y recién entonces ejecuta el cuerpo.

Python lee el archivo de arriba hacia abajo. Por eso una función debe estar definida antes de la primera llamada que queramos ejecutar.

```python
# Objetivo: observar qué ocurre al definir una función y qué ocurre al llamarla.
print("1. Antes de definir la función")

def calcular_doble(numero):
    print("3. Ahora se está ejecutando el cuerpo")
    return numero * 2

print("2. La función ya fue definida, pero su cuerpo todavía no se ejecutó")
resultado = calcular_doble(6)
print("4. Resultado recibido:", resultado)

# Conclusión: la definición guarda el cuerpo; la llamada es la que lo ejecuta.
```

**Una consecuencia importante:**

Si escribimos una llamada antes de que Python haya leído la definición, el nombre de la función todavía no existe y se produce un `NameError`. En un notebook también debemos ejecutar primero la celda que contiene `def`.

## 4. Entregar resultados con `return`

`return` entrega un resultado al lugar desde donde llamamos a la función. Ese resultado puede guardarse en una variable, imprimirse, compararse o participar en otra operación.

Una función no está obligada a usar `return`. Si termina sin encontrar uno, devuelve `None`, que representa la ausencia de un resultado.

```python
# Objetivo: retornar valores y utilizarlos fuera de la función.
def calcular_total(precio_unitario, cantidad):
    total = precio_unitario * cantidad
    return total

total_cuadernos = calcular_total(2500, 3)
total_lapices = calcular_total(800, 5)
total_compra = total_cuadernos + total_lapices

print("Total de cuadernos:", total_cuadernos)
print("Total de lápices:", total_lapices)
print("Total de la compra:", total_compra)

# Conclusión: return permite sacar un resultado de la función y reutilizarlo
# en cálculos posteriores.
```

### 4.1 `return` no es lo mismo que `print()`

`print()` muestra información en pantalla. `return` entrega un valor para que el resto del programa pueda usarlo.

Una función que solo imprime puede ser útil para mostrar un mensaje, pero lo mostrado no queda disponible como resultado de la función.

```python
# Objetivo: comparar una función que imprime con una función que retorna.
def mostrar_cuadrado(numero):
    print("Cuadrado mostrado:", numero * numero)

def calcular_cuadrado(numero):
    return numero * numero

resultado_mostrado = mostrar_cuadrado(5)
resultado_retornado = calcular_cuadrado(5)

print("Valor retornado por mostrar_cuadrado:", resultado_mostrado)
print("Valor retornado por calcular_cuadrado:", resultado_retornado)

# Conclusión: mostrar_cuadrado imprime 25, pero retorna None. En cambio,
# calcular_cuadrado retorna 25 y permite guardarlo.
```

### 4.2 `return` también termina la ejecución

Cuando Python llega a `return`, entrega el resultado y sale inmediatamente de la función. Las instrucciones que estén después dentro de ese mismo camino ya no se ejecutan.

Podemos aprovechar este comportamiento para resolver un caso y terminar sin revisar condiciones innecesarias.

```python
# Objetivo: usar return para terminar la función cuando un caso ya está resuelto.
def clasificar_nota(nota):
    if nota < 1 or nota > 7:
        return "Nota fuera del rango permitido"
    if nota >= 4:
        return "Aprobada"
    return "No aprobada"

print("Nota 6.1:", clasificar_nota(6.1))
print("Nota 3.5:", clasificar_nota(3.5))
print("Nota 8.0:", clasificar_nota(8.0))

# Conclusión: cada llamada termina en el primer return que corresponde a su caso.
```

## 5. Alcance o *scope*

El **alcance** indica en qué parte del programa está disponible un nombre:

- Una variable creada fuera de una función pertenece al alcance **global**.
- Los parámetros y las variables creadas dentro de una función pertenecen a su alcance **local**.
- Cada llamada crea un alcance local nuevo. Al terminar la llamada, sus variables locales dejan de estar disponibles.
- `return` permite entregar hacia afuera el valor de una variable local.

Una función puede leer variables globales, pero suele ser más claro entregarle la información mediante parámetros. Así sabemos exactamente qué datos necesita y podemos reutilizarla con otros valores.

```python
# Objetivo: distinguir un nombre global de los nombres locales de una función.
tema = "funciones propias"  # Variable global.

def crear_resumen(tema):  # Este parámetro local también se llama tema.
    mensaje = "El próximo tema será: " + tema  # Variable local.
    print("Tema dentro de la función:", tema)
    return mensaje

resumen = crear_resumen("listas")
print("Tema fuera de la función:", tema)
print("Resumen retornado:", resumen)

'''
Conclusión: dentro de la función, tema vale "listas". Fuera de ella, la variable
global tema conserva "funciones propias". mensaje sale únicamente mediante return.
'''
```

### 5.1 Cada llamada tiene sus propias variables locales

Las variables locales no recuerdan automáticamente su valor entre una llamada y la siguiente. Cada ejecución comienza con un alcance local nuevo.

```python
# Objetivo: comprobar que una variable local vuelve a crearse en cada llamada.
def avanzar_una_vez():
    contador = 0
    contador = contador + 1
    return contador

print("Primera llamada:", avanzar_una_vez())
print("Segunda llamada:", avanzar_una_vez())

# Conclusión: ambas llamadas retornan 1 porque cada una crea su propio contador
# local y comienza desde 0.
```

**Errores comunes relacionados con el alcance:**

- Intentar usar fuera de la función una variable que fue creada dentro de ella.
- Pensar que un parámetro cambia automáticamente una variable global con el mismo nombre.
- Depender de muchas variables globales ocultas. Entregar los datos como parámetros hace que la función sea más fácil de entender y probar.

## 6. Argumentos por nombre o *keyword arguments*

Hasta ahora hemos entregado argumentos según su posición. En `crear_ficha("Ana", 20, "Santiago")`, el primer valor corresponde al primer parámetro, el segundo al segundo y el tercero al tercero.

También podemos escribir el nombre del parámetro en la llamada: `crear_ficha(nombre="Ana", edad=20, ciudad="Santiago")`. Estos son **argumentos por nombre** o **argumentos por palabra clave**.

Al usar nombres:

- La llamada explica claramente qué representa cada valor.
- Podemos cambiar el orden de los argumentos.
- El nombre escrito debe coincidir exactamente con un parámetro.

```python
# Objetivo: comparar argumentos por posición con argumentos por nombre.
def crear_ficha(nombre, edad, ciudad):
    return nombre + " tiene " + str(edad) + " años y vive en " + ciudad + "."

ficha_posicional = crear_ficha("Ana", 20, "Santiago")
ficha_por_nombre = crear_ficha(ciudad="Santiago", nombre="Ana", edad=20)

print("Con argumentos por posición:", ficha_posicional)
print("Con argumentos por nombre:", ficha_por_nombre)

# Conclusión: ambas llamadas producen el mismo resultado. Los nombres permiten
# cambiar el orden sin cambiar qué valor recibe cada parámetro.
```

**Errores comunes:**

- Inventar un nombre que no aparece en la definición. `crear_ficha(persona="Ana", ...)` falla porque el parámetro se llama `nombre`.
- Entregar dos veces el mismo parámetro, una por posición y otra por nombre.
- Colocar un argumento por posición después de uno por nombre. En las llamadas que mezclan ambos estilos, los argumentos por posición deben ir primero.

## 7. Parámetros por defecto

Un parámetro puede tener un valor preparado en la definición. Ese valor se usa cuando la llamada no entrega otro.

En `def saludar(nombre, saludo="Hola"):`, `nombre` es obligatorio y `saludo` tiene el valor por defecto `"Hola"`. Podemos omitir `saludo` o reemplazarlo en una llamada.

Los parámetros obligatorios se escriben antes de los parámetros que tienen un valor por defecto.

```python
'''
Estructura general:

def nombre_funcion(parametro_obligatorio, parametro_opcional=valor_por_defecto):
    instrucciones de la función
'''

# Objetivo: usar valores por defecto y reemplazarlos mediante argumentos por nombre.
def crear_saludo(nombre, saludo="Hola", cierre="!"):
    return saludo + ", " + nombre + cierre

print(crear_saludo("Camila"))
print(crear_saludo("Camila", saludo="Buenos días"))
print(crear_saludo("Camila", cierre=".", saludo="Bienvenida"))

# Conclusión: los valores por defecto simplifican el caso habitual, mientras
# los argumentos por nombre permiten reemplazar solo lo necesario.
```

### 7.1 Definición y ejecución de los valores por defecto

El valor por defecto queda asociado al parámetro cuando Python define la función. En cada ejecución, Python revisa si la llamada entregó ese argumento:

- Si lo entregó, usa el argumento de la llamada.
- Si lo omitió, usa el valor por defecto.

Para comenzar, conviene usar valores simples como números, textos, `True`, `False` o `None`. Las listas y otros objetos que pueden modificarse requieren un cuidado adicional cuando se usan como valores por defecto.

## 8. Ejemplo integrador

El siguiente ejemplo reúne los conceptos principales:

- Parámetros obligatorios y por defecto.
- Argumentos por nombre.
- Variables locales.
- Un valor entregado mediante `return`.
- Varias ejecuciones de una misma definición.

```python
# Objetivo: combinar parámetros, valores por defecto, alcance local y return.
def calcular_compra(precio_unitario, cantidad=1, descuento=0):
    subtotal = precio_unitario * cantidad
    monto_descuento = subtotal * descuento
    total = subtotal - monto_descuento
    return total

compra_individual = calcular_compra(2500)
compra_multiple = calcular_compra(2500, cantidad=3)
compra_con_descuento = calcular_compra(
    precio_unitario=2500,
    cantidad=3,
    descuento=0.10
)

print("Una unidad:", compra_individual)
print("Tres unidades:", compra_multiple)
print("Tres unidades con 10% de descuento:", compra_con_descuento)

'''
Conclusión: una sola definición resuelve tres casos. subtotal, monto_descuento
y total son variables locales; return entrega únicamente el resultado final.
'''
```

## 9. Resumen

- `def` crea una función; `nombre_funcion()` la ejecuta.
- Los parámetros aparecen en la definición y los argumentos aparecen en la llamada.
- Definir guarda las instrucciones; llamar ejecuta el cuerpo.
- `return` entrega un valor y termina la ejecución de la función.
- Los parámetros y las variables creadas dentro de la función tienen alcance local.
- Los argumentos por nombre indican explícitamente qué parámetro recibe cada valor.
- Los parámetros por defecto permiten omitir argumentos en los casos habituales.
