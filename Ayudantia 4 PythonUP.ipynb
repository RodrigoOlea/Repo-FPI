# **Funciones nativas**, **Funciones importadas** y **Funciones propias** en Python


## 1) Que es una *funcion importada*?
- Una **función importada** es una funcion que proviene de otro archivo (un **módulo** o **paquete**) y se ***llama*** con la palabra reservada `import`.
- Se usa mucho para emplear soluciones mas sofisticadas y/o completas, ya que un **modulo** puede tener una variedad de herramientas utiles para un tipo especifico de tareas.
- La **biblioteca estándar** de Python *(es decir, el conjunto de modulos que podemos importar en Python, sin tener que instalar nada adicional)* ya incluye algunos módulos útiles: `math`, `random`, `statistics`, `datetime`, `os`, etc.


```python
# Ejemplo: el modulo math incluye herramientas matematicas, y podemos acceder a ellas importandolo antes de usarlas.
import math

print("√25 =", math.sqrt(25))
print("pi =", math.pi)
```

## 2) Como importamos un modulo?

### 2.1. **Importando el modulo completo:** `import <modulo>`
Podemos importar el modulo completo, y esto nos da acceso a todas las funciones que el modulo ofrece. Sin embargo, cuando importamos las funciones de esta forma, **todas las funciones del modulo deben llevar el nombre del modulo al principio**, ya que este funciona como un identificador.

```python
import math # import <modulo>
print(math.sqrt(16)) # Vemos que la funcion "sqrt" debe llamarse como "math.sqrt()"
print(math.ceil(3.1)) 
print(math.pi)
# Para llamar a cualquier funcion en un modulo de esta manera, en general se seguira el formato: <modulo>.<funcion>()
```

Al importar el modulo de esta forma, siempre sera necesario llamar a las funciones importadas con su respectivo identificador. Imaginemos que el modulo math incluyera una funcion llamada "sum()". ¿Como podriamos esperar que Python sepa si nos referimos al "sum()" del modulo math, o el "sum()" nativo de Python?

Es por ese motivo que debemos usar el identificador, ya que Python necesita evitar cualquier posible conflicto con el nombre de las funciones, ya que una funcion importada puede generar conflicto con una nativa.

### 2.2 `from modulo import funcion1, funcion2, ...`
Este formato nos permite importar individualmente las funciones que necesitemos de un modulo. Dado que al importar funciones con este metodo sabemos los nombres de todas las funciones que estamos importando, Python nos da la libertad de no tener que usar el identificador del modulo para llamar la funcion, y nos permite llamarlas tal como llamariamos a una funcion nativa cualquiera.

```python
# Usamos la palabra reservada "from" (desde) para obtener funciones en especifico `desde` el modulo
from math import sqrt, pi # from <modulo> import <funcion1>, <funcion2>, ... 
print(sqrt(49))
print(pi)
# Al importarlas de esta forma las funciones se llaman como cualquier otra funcion nativa.
```

### 2.3 `import modulo as apodo`
Este metodo es esencialmente igual al primero, en el sentido de que debemos usar un identificador para llamar a las funciones. La diferencia es que podemos darle un apodo de preferencia, si se vuelve molesto escribir reiteradamente el modulo, por ejemplo.


```python
# Usamos la palabra reservada "as" (como) para asignarle un apodo al modulo.
import statistics as stats # import <modulo> as <apodo>
print(stats.mean([1, 2, 10])) # Ahora usaremos el apodo en vez de el identificador original para llamar a las funciones del modulo.
```

### 2.4 `from modulo import *` (evitar)
Este metodo es bastante susceptible a conflictos, ya que importa **todas las funciones y elimina la necesidad del identificador**.

Es preferible evitar su uso, ya que se presta bastante a errores; si por algun casual algun nombre genera conflicto con una funcion nativa de Python, la forma de solucionarlo sera simplemente usar otro metodo de importacion si queremos usar dicha funcion importada.

## 3) **Funciones propias:**
Las **funciones propias** son funciones *creadas por el usuario mismo*. Para ***definir*** una funcion utilizamos la palabra reservada `def`. La capacidad de definir una funcion propia es muy importante: nos permite, entre otras cosas, generalizar una serie de instrucciones, lo que puede ayudarnos a hacer nuestro codigo mas conciso y legible, incluso mas comodo de escribir y de expandir

Hay una variedad de reglas y convenciones alrededor de la tarea de escribir nuestra propia funcion, por lo que las trabajaremos una a una.

### 3.1) **Sintaxis y estructura**
La creacion de una funcion sigue una estructura *(relativamente)* fija, de Encabezado, Cuerpo, y Salida.


```python
def calcular_sumatoria(numero): # ENCABEZADO
    ### CUERPO ###
    i = 0
    suma = 0
    while i <= numero:
        suma = suma + i
        i = i + 1
    ##############
    return suma # SALIDA
```

#### ***Encabezado:***
El encabezado de una funcion es la linea que engloba todo el set de instrucciones que ejecutará la función. En general, sigue la sintaxis: `def <nombre_funcion>(<parametros>)`. Es ideal que el nombre de la funcion sea descriptivo y comodo de escribir, ya que el proposito de escribir la funcion en primer lugar es llamarla mas adelante, probablemente reiteradas veces, dentro de tu codigo.

Por otro lado, los `parametros` de la funcion, que van dentro del parentesis, son ***todos los datos pertinentes que requiere la funcion para cumplir su cometido***. Si el proposito de mi funcion fuera, por ejemplo, calcular el area de un rectangulo, debo entregarle como minimo, la longitud de ambos lados de la figura. Sin embargo, para **decidir que parametros entregarle a la funcion**, primero debemos revisar el concepto de ***alcance*** o ***scope*** de la funcion

##### ***Qué es el scope de mi funcion?*** 
Con ***scope*** (o alcance) nos referimos al concepto de que **una variable solo esta disponible dentro de la region en la que fue creada, por lo que tiene un cierto alcance, vive dentro de un definido *scope*.**

El scope con el que estamos acostumbrados a trabajar es el scope **global** de Python. Cuando declaro una variable en el cuerpo principal de mi codigo, puedo volver a acceder a esta variable mas abajo sin problema. Sin embargo, ya que una funcion propia busca ser **una serie de instrucciones genericas**, cuando declaramos una funcion, estamos trabajando en un alcance distinto, donde ***no podemos acceder a las cosas que estan en nuestro scope global, ni viceversa*.**

```python
x = 10 # La variable `x` esta declarada en el scope global, la puedo llamar en cualquier parte sin problema.
print(10*x)

def funcion_de_ejemplo():
    print(x) # No puedo utilizar la variable `x` dentro de la funcion, ya que no existe en el scope de la funcion
    y = 5 # Sin embargo, al declarar `y` dentro de la funcion, la variable solo existe dentro de la funcion, y no es posible llamarla por fuera
    print(10*y)

#Python ni siquiera reconoce la variable `y`, ya que no existe dentro de nuestro scope.
#print(y)
```

Ahora que entendemos que las funciones son su propio mundo debido a su scope, podemos entender la necesidad que cubren los **parametros**; los parametros que le entregamos a la funcion son ***la ventana para poder entregarle variables del scope global a la funcion*.** Si volvemos al ejemplo inicial de la sumatoria, podemos ver esta relacion. Todo lo que necesitamos saber para calcular la sumatoria desde 0 a algun numero, es dicho numero. Por lo tanto, es conocimiento suficiente para que la funcion trabaje. Nuevamente, los parametros cumplen el proposito de entregarle el conocimiento suficiente a la funcion para que esta pueda cumplir su cometido.

#### ***Cuerpo:***
El cuerpo de la funcion es la mas auto-descriptiva, hay que usar los parametros entregados para obtener la salida deseada. Sin embargo, aun queda algo por decir acerca de como escribimos la **salida** de una funcion.

#### ***Salida***
La salida se caracteriza por la palabra reservada `return`. Cuando queremos arrojar el resultado de nuestra funcion, ***retornamos*** el resultado. Sin embargo, esto no se debe confundir con **imprimir** el resultado. **(return != print)!!!**

```python
def numero_al_cuadrado(numero):
    salida = numero*numero
    return salida

numero_al_cuadrado(5) # Llamar por si sola la funcion `numero_al_cuadrado()` no imprime nada en absoluto, ya que no posee ninguna instruccion de print.
# La salida que retornamos gracias a `return` es efectivamente el resultado de la funcion numero_al_cuadrado. Sin embargo, llamarla por si sola no hace nada con ese resultado,
# no ha sido impresa con print, o usada en una operacion. La funcion por si sola no es NADA mas que la salida retornada, y depende de nosotros lo que hacemos con esa salida.

print(numero_al_cuadrado(5)) # Ahora si, imprimimos el retorno de la funcion
operacion = numero_al_cuadrado(5) * numero_al_cuadrado(10) # O incluso podemos usarla en una operacion
print(operacion) # E imprimir el resultado
```
